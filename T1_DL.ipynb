**Justificación de Arquitectura**

La red está separada en dos partes: una parte convolucional y un clasificador. En la primera, notamos que el número de parámetros y de filtros va aumentando conforme se avanza en las capas. Esto tiene sentido, pues se parte de la idea de que las primeras capas deben aprender solamente conceptos fundamentales, por lo cual no necesitan demasiada variabilidad. Sin embargo, a medida que se va avanzando en las capas, se van realizando conjugaciones de estos distintos conceptos básicos aprendidos, por lo cual se necesita mayor grados de libertad en estas conjugaciones. Esto es similar al ejemplo del alfabeto occidental: es necesario partir con un conjunto limitado de letras (entre 26 y 28), es decir, con pocos grados de libertad, a partir de las cuales se pueden ir generando una cantidad exponencial de frases, oraciones, parrafos, etc. cada vez mas complejos (mayores grados de libertad).

## Parte 2.1: Actividad 8

A continuación, definen funciones que serán utilizadas a lo largo de los experimentos:

```python
# Funciones auxiliares

## load_data: Carga la data e infiere las clases de manera eficiente

def load_data(path, batch_size, epochs, input_size):
  train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2, # random application of shearing
    zoom_range = 0.2,
    horizontal_flip = True) 
  
  test_datagen = ImageDataGenerator(rescale = 1./255)
  
  
  train_generator = train_datagen.flow_from_directory(
    directory=train_path,
    target_size=input_size,
    color_mode="rgb",
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True,
    seed=42
)
  
  validation_generator = test_datagen.flow_from_directory(
    directory=validation_path,
    target_size=input_size,
    color_mode="rgb",
    batch_size=batch_size,
    class_mode="categorical",
    shuffle=True,
    seed=42
)
  
  test_generator = test_datagen.flow_from_directory(
    directory=test_path,
    target_size=input_size,
    color_mode="rgb",
    batch_size=batch_size,
    class_mode=None,
    shuffle=False,
    seed=42
)


  return train_generator, validation_generator, test_generator

# modelo_resnet50: Carga y compila el modelo de red convolucional.

def modelo_resnet50(perdida, optim, num_classes, input_size):
  modelo = ResNet50(weights=None, input_shape=input_size, classes=num_classes)
  modelo.compile(loss=perdida, optimizer=optim , metrics=['accuracy'])
  return modelo


# entrenar_modelo : Realiza el entrenamiento y guardado de información (modelo e historia)
def entrenar_modelo(modelo, train_generator, validation_generator, batch_size, epochs, checkpoint_path, save_path, history_path):
  
  history = History()
  cp_callback = ModelCheckpoint(checkpoint_path,
                                save_weights_only=False,
                                period=5,
                                verbose=1)
  
  early_stopping = EarlyStopping(monitor='val_loss', min_delta=0.001, patience=5, verbose=1, mode='auto')

  
  a = modelo.fit_generator(train_generator, 
                 steps_per_epoch= train_generator.n//train_generator.batch_size, 
                 epochs=epochs, 
                 validation_data=validation_generator, 
                 validation_steps=validation_generator.n//validation_generator.batch_size,
                       verbose=1,
                       callbacks = [history, cp_callback, early_stopping])

  
  modelo.save(save_path)
  history_dict = history.history
  # Save it under the form of a json file
  json.dump(history_dict, open(history_path, 'w'))
  
  return (modelo, history, a)
  


# load_modelo_full: Carga un modelo previamente almacenado.
def load_modelo_full(path):
  modelo = keras.models.load_model(path)
  return modelo



# plot_history: Grafica la Evolución de la función de pérdida y la precisión

def plot_history(history, epochs):

  fig, axs = pyplot.subplots(1, 2)

  axs[0].plot(np.arange(0, epochs,1)+1, history['loss'], label='Train Loss', color='green', marker='o', linestyle='dashed')
  axs[0].plot(np.arange(0, epochs,1)+1, history['val_loss'], label='Validation Loss', color='blue', marker='o', linestyle='dashed')


  axs[1].plot(np.arange(0, epochs,1)+1, history['acc'], label='Train Accuracy', color='green', marker='o', linestyle='dashed')
  axs[1].plot(np.arange(0, epochs,1)+1, history['val_acc'], label='Validation Accuracy', color='blue', marker='o', linestyle='dashed')



  axs[0].set_title('Loss: Train vs Validation')

  axs[1].set_title('Acurracy: Train vs Validation')

  for ax in axs.flat:
      ax.set(xlabel='Epochs', ylabel='Value')
      ax.legend()

```

   * Para el tamaño de minibatch, se probó con una grilla de valores y luego se escogió aquella que tardó menor tiempo y presentó resultados más estables (precisión y pérdida).
```python
fig2, axs2 = pyplot.subplots(ncols=3, nrows=3)
fig2.suptitle("Clase Mejor Evaluada: {0}".format(p2))
for i in range(3):
  for j in range(3):
    img = cv2.imread(path_best+best_images[3*i+j], cv2.IMREAD_UNCHANGED)
    axs2[i, j].imshow(img)
    
```
